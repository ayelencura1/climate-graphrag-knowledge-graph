## 3. Run a Deep Research GRAPH Query (the Retrieval)


  We will now execute a comprehensive investigation using a hybrid retrieval architecture that combines semantic understanding with structural reasoning. To control the scope and precision of this process, the
  engine exposes three critical hyperparameters that govern how information is extracted from the "Brain."

  The first parameter, `top_k`, establishes the Breadth of the investigation within the semantic vector space. This controls how many distinct conceptual "anchors" or graph nodes the engine identifies initially.
  By scanning the entire high-dimensional space, the system locates the concepts that are mathematically closest to your query, effectively casting a wide net to capture the most relevant primary topics.

  Once these anchor topics are established, the `context_k` parameter dictates the Depth of the analysis. Instead of relying on a single document per topic, the engine performs a secondary, localized vector search
  within each identified node. This retrieves additional, highly relevant papers or abstracts specific to that concept, ensuring that the final response is grounded in a robust density of evidence rather than an
  isolated data point.

  Finally, the `hops` parameter activates the Causal Reasoning layer. Unlike the previous steps which rely on text similarity, this stage traverses the topology of the Knowledge Graph. By setting this to 2, you
  instruct the engine to follow the connections (edges) two steps away from the source. This is essential for uncovering "second-order effects"â€”cascading risks where an initial event (like a heatwave) triggers a
  chain reaction that impacts downstream systems (like food security) through intermediate causes.


    Summary:
   * Vector Search (`top_k`) finds concepts.
   * Context Expansion (`context_k`) builds knowledge about those concepts.
   * Graph Traversal (`hops`) discovers the hidden stories connecting them.



```python
# Define your research question
query = "Which vulnerable social groups are most severely affected by extreme heat, and what specific socioeconomic factors exacerbate their exposure?"

# Execute the Search
results = engine.search(
    query=query,
    top_k=5,        # Breadth
    context_k=3,    # Depth
    hops=2          # Causality
)

print(f"--- Research Stats ---")
print(f"Primary Sources: {results['stats']['primary']}")
print(f"Context Expansion: {results['stats']['context']}")
print(f"Causal Links:      {results['stats']['graph']}")
```

fijense las Stats para tener una idea rapida de que clase de contexto se usara. Si es demasiado, o poco, ajustar con los parametros.

To replicate a traditional RAG (Retrieval-Augmented Generation) workflow, effectively bypassing the graph's structural capabilities, one would configure the system with `hops=0` and `context_k=0`.

  In this specific configuration, the engine relies exclusively on the `top_k` parameter to perform a standard vector similarity search across the entire document corpus. By setting hops to zero, the system is
  prevented from traversing any causal edges, thereby ignoring the relational topology entirely. Similarly, setting context_k to zero disables the node-centric context expansion, ensuring that only the single
  best-matching snippet for each result is retrieved. This setup is particularly useful for benchmarking, as it isolates the performance of pure semantic search against the enhanced, graph-aware methodology.
