```python
import json
from collections import defaultdict
from pathlib import Path
from copy import deepcopy


def load_graph(path: Path) -> dict:
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)


def merge_graphs(graph_files):
    merged_nodes = {}
    merged_edges = {}
    connection_counter = defaultdict(int)

    for graph_file in graph_files:
        graph = load_graph(graph_file)

        # ---------- MERGE NODES ----------
        for node in graph.get("nodes", []):
            node_id = node["id"]

            if node_id not in merged_nodes:
                merged_nodes[node_id] = {
                    "id": node_id,
                    "label": node.get("label"),
                    "type": node.get("type"),
                    "properties": deepcopy(node.get("properties", {})),
                    "_validationStatus": node.get("_validationStatus", "ok"),
                    "connections": 0
                }
            else:
                if node.get("_validationStatus") == "error":
                    merged_nodes[node_id]["_validationStatus"] = "error"

        # ---------- MERGE EDGES ----------
        for edge in graph.get("edges", []):
            src = edge["source"]
            tgt = edge["target"]
            rel = edge.get("relation")
            weight = edge.get("weight", 1)

            edge_key = (src, tgt, rel)

            if edge_key not in merged_edges:
                merged_edges[edge_key] = {
                    "source": src,
                    "target": tgt,
                    "relation": rel,
                    "weight": weight
                }
            else:
                merged_edges[edge_key]["weight"] += weight

    # ---------- CONNECTIONS ----------
    for edge in merged_edges.values():
        connection_counter[edge["source"]] += 1
        connection_counter[edge["target"]] += 1

    for node_id, node in merged_nodes.items():
        node["connections"] = connection_counter.get(node_id, 0)

    return {
        "nodes": list(merged_nodes.values()),
        "edges": list(merged_edges.values())
    }


# ====== ESTA PARTE ES LA CLAVE EN COLAB ======
graphs_dir = Path("/content")              # ← aquí están los JSON
graph_files = list(graphs_dir.glob("*.json"))

if not graph_files:
    raise ValueError("No se encontraron archivos JSON en /content")

merged_graph = merge_graphs(graph_files)

with open("merged_graph.json", "w", encoding="utf-8") as f:
    json.dump(merged_graph, f, indent=2, ensure_ascii=False)

print(f"Merged {len(graph_files)} graphs")
print(f"Total nodes: {len(merged_graph['nodes'])}")
print(f"Total edges: {len(merged_graph['edges'])}")
```

```python
import json
from collections import defaultdict
from pathlib import Path
from copy import deepcopy


def load_graph(path: Path) -> dict:
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)


def merge_graphs(graph_files):
    merged_nodes = {}
    merged_edges = {}
    connection_counter = defaultdict(int)

    for graph_file in graph_files:
        graph = load_graph(graph_file)

        # ---------- MERGE NODES ----------
        for node in graph.get("nodes", []):
            node_id = node["id"]

            if node_id not in merged_nodes:
                merged_nodes[node_id] = {
                    "id": node_id,
                    "label": node.get("label"),
                    "type": node.get("type"),
                    "properties": deepcopy(node.get("properties", {})),
                    "_validationStatus": node.get("_validationStatus", "ok"),
                    "connections": 0
                }
            else:
                # Merge validation status
                if node.get("_validationStatus") == "error":
                    merged_nodes[node_id]["_validationStatus"] = "error"

                # Merge references inside properties
                new_refs = node.get("properties", {}).get("references")
                if new_refs:
                    props = merged_nodes[node_id]["properties"]

                    if "references" not in props:
                        props["references"] = []

                    # Evitar duplicados (opcional pero recomendable)
                    existing = {
                        (r.get("source_id"), r.get("origin"))
                        for r in props["references"]
                    }

                    for ref in new_refs:
                        key = (ref.get("source_id"), ref.get("origin"))
                        if key not in existing:
                            props["references"].append(ref)

        # ---------- MERGE EDGES ----------
        for edge in graph.get("edges", []):
            src = edge["source"]
            tgt = edge["target"]
            rel = edge.get("relation")
            weight = edge.get("weight", 1)

            edge_key = (src, tgt, rel)

            if edge_key not in merged_edges:
                merged_edges[edge_key] = {
                    "source": src,
                    "target": tgt,
                    "relation": rel,
                    "weight": weight
                }
            else:
                merged_edges[edge_key]["weight"] += weight

    # ---------- CONNECTIONS ----------
    for edge in merged_edges.values():
        connection_counter[edge["source"]] += 1
        connection_counter[edge["target"]] += 1

    for node_id, node in merged_nodes.items():
        node["connections"] = connection_counter.get(node_id, 0)

    return {
        "nodes": list(merged_nodes.values()),
        "edges": list(merged_edges.values())
    }


# ====== ESTA PARTE ES LA CLAVE EN COLAB ======
graphs_dir = Path("/content")              # ← aquí están los JSON
graph_files = list(graphs_dir.glob("*.json"))

if not graph_files:
    raise ValueError("No se encontraron archivos JSON en /content")

merged_graph = merge_graphs(graph_files)

with open("merged_graph.json", "w", encoding="utf-8") as f:
    json.dump(merged_graph, f, indent=2, ensure_ascii=False)

print(f"Merged {len(graph_files)} graphs")
print(f"Total nodes: {len(merged_graph['nodes'])}")
print(f"Total edges: {len(merged_graph['edges'])}")
```
